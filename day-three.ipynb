```python
password = (input("Enter your password: "))

length = False
digit = False
symbol = False
lower = False
upper = False
symbols = "!@#$%^&*()-_+={}[]:;'<>,./?*`~"

for ch in password:
    if ch.isdigit():
        digit = True
    if ch.islower():
        lower = True
    if ch.isupper():
        upper = True
    if ch in symbols:
        symbol = True
if len(password) >= 8:
    length = True
    print ("Password Strength")
if length and digit and symbol and lower and upper:
    print ("Strong Password")
elif length and (digit or symbol or lower or upper):
    print ("Medium Password")
else:
    print ("Weak Password")
```
